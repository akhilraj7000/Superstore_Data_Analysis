## Removing unwanted column(s)

```python
df.drop(['Row_ID', 'Country'], axis = 1, inplace = True)
df.head(2)
```

## Removing duplicate rows ( there could be duplicate rows excluding Row_ID column ).

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace = True, ignore_index = True)
df.head(2)
```

```python
df.shape
```

## Removing rows for which few values are missing.

```python
df.isna().sum()
```

```python
df.dropna(inplace = True)
df.head(2)
```

```python
df.shape
```

## Remove irrelevant values from each column if any. Validation of all values for a column( order date and ship date value must be in correct date format ). For each entry in dataset ship date >= order date

### Transforming to correct datatype

```python
df.info()
```

```python
df = df.astype({'Order_Date': 'datetime64[ns]', 'Ship_Date': 'datetime64[ns]', 'Postal_Code': 'int'})
df['Postal_Code'] = df['Postal_Code'].map(str)
df.info()
```

### Checking weather Ship_Date < Order_Date

```python
# Ship_Date < Order_Date -> 1684 rows
df.loc[df['Ship_Date'] < df['Order_Date']].head(2)
```

```python
df['Ship_Mode'].value_counts()
```

```python
# Error when the 'Ship_Mode' is 'Second Class' -> 305 rows
df.loc[(df['Ship_Mode'] == 'Second Class') & (df['Ship_Date'] < df['Order_Date'])].head(2)
```

```python
# Error when the 'Ship_Mode' is 'First Class' -> 157 rows
df.loc[(df['Ship_Mode'] == 'First Class') & (df['Ship_Date'] < df['Order_Date'])].head(2)
```

```python
# Error when the 'Ship_Mode' is 'Same Day' -> 2 rows
df.loc[(df['Ship_Mode'] == 'Same Day') & (df['Ship_Date'] < df['Order_Date'])]
```

```python
# Error when the 'Ship_Mode' is 'Standard Class' -> 1220 rows
df.loc[(df['Ship_Mode'] == 'Standard Class') & (df['Ship_Date'] < df['Order_Date'])].head(2)
```

### Removing rows where 'Ship_Date' < 'Order_Date'

```python
df = df.loc[df['Ship_Date'] >= df['Order_Date']]
df.shape
```

```python
df.reset_index(drop = True, inplace = True)
df.head(2)
```

## Export the cleaned dataset as a .csv file: prefer UTF-8 encoding.

```python
df.to_csv('Superstore_v2.csv', encoding='utf-8')
```
